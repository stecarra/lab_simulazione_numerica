### <span style="color:green"> Equilibration is needed!</span>

By monitoring the temperature, $T$, during the first time steps of the simulation, you will observe that at the beginning, starting from a perfect crystalline configuration, $T$ will not conserve its value. Can you understand and explain this behaviour?

Explore and find the starting value of $T$ needed to stabilize, after a certain number of time steps (equilibration time), the average value of $T$ on the desired $T^\star$. <span style="color:red">Perform your solid/liquid phase MD simulations by re-starting them after such equilibration time.</span>

Liquid and solid phase stabilize about the thermal equilibrium moree rapidly than the gas phase, which spends more than 10 times to reach it.
The reason the simulation need some equilibration steps if started from a crystalin configuration is that this configuration is very particular: it minimize the potential energy. 



Since the temperature we set at the beginning is used to distribute the initial velocities and thus the initial kinetic energies, and since in this simulation total energy is conserved, the potential energy will change value to a higher one (as it gets away from minimum) while the velocities will decrease, leading to a general decrease in temperature. It is for this reason that to reach a fixed equilibrium temperature we ought to set an initial higher one.
